```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
# genrating random data 
X= -2 * np.random.rand(100,2)
X1 = 1 + 2 * np.random.rand(50,2)
X[50:100, :] = X1
plt.scatter(X[ : , 0], X[ :, 1], s = 50)
plt.show()
```

```python
Kmean = KMeans(n_clusters=2)
Kmean.fit(X)
```

```python
Kmean.cluster_centers_
```

```python
plt.scatter(X[ : , 0], X[ : , 1], s =50, c='b')
plt.scatter(-1.04821957, -1.03747706, s=200, c='g', marker='s')
plt.scatter(2.03482252, 1.96927039, s=200, c='r', marker='s')
plt.show()
```

```python
# using feature scaling 
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
weights = np.array([[115.],[140.],[175.]])
new_weights = scaler.fit_transform(weights)
print(new_weights)
```

```python
new_kmean = scaler.fit_transform(X)
```

```python
Kmean = KMeans(n_clusters=2)
Kmean.fit(new_kmean)
Kmean.cluster_centers_
```

```python
plt.scatter(new_kmean[ : , 0], new_kmean[ : , 1], s =50, c='b')
plt.scatter(0.19067365, 0.18811725, s=200, c='g', marker='s')
plt.scatter(0.8084357 , 0.79420819, s=200, c='r', marker='s')
plt.show()
```
